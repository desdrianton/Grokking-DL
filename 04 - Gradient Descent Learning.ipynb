```python
MAX_ITERATION = 100
MINUTES_RUNNING = 20
ACTUAL_CALORIES_BURNED = 180
INITIAL_WEIGHT = 7

SLOW_LEARNING_RATE = 0.01
OSCILLATING_LEARNING_RATE = 0.1

learning_rate = SLOW_LEARNING_RATE


def neural_network(input, weight):
    predicted_value = input * weight
    return predicted_value


input = MINUTES_RUNNING
expected_value = ACTUAL_CALORIES_BURNED
weight = INITIAL_WEIGHT
iteration = 1

while True:
    prediction = neural_network(input, weight)
    error = (prediction - ACTUAL_CALORIES_BURNED) ** 2
    derivative = 2 * (prediction - expected_value)
    
    print(f"=================== Iteration: {iteration}")
    print(f"weight            : {weight}")
    print(f"prediction        : {prediction}")
    print(f"error             : {error}")
    print(f"error (round)     : {round(error, 5)} {"     ===> Yeayyyy the error is zero !!!!" if round(error, 5) == 0 else ""}")
    print(f"derivative        : {derivative}")

    weight_adjustment = learning_rate * derivative
    weight -= weight_adjustment

    print(f"weight_adjustment : {weight_adjustment}")
    print(f"weight (new)      : {weight}")
    print()

    iteration += 1
    if(round(error, 5) == 0 or iteration > MAX_ITERATION):
        break
```
