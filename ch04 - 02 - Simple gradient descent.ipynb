```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.01
iteration = 0
do_iteration = True

while do_iteration:
    prediction = input * weight
    error = round((prediction - goal_prediction) ** 2,5)
    direction_and_amount = (prediction - goal_prediction) * input

    print(f"Iteration {iteration}")
    print(f"- weight: {weight}")
    print(f"- Prediction: {prediction}; Error: {error};")
    

    weight = weight - direction_and_amount
    do_iteration = (error != 0.00000)
    iteration += 1
    
    print(f"- new weight: {weight}")
    print()
```
