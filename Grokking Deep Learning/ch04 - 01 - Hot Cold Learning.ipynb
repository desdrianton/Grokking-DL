```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.01
iteration = 0
do_iteration = True

while do_iteration:
    prediction = input * weight
    error = round((prediction - goal_prediction) ** 2,5)

    up_prediction = input * (weight + step_amount)
    up_error = round((goal_prediction - up_prediction) ** 2,5)

    down_prediction = input * (weight - step_amount)
    down_error = round((goal_prediction - down_prediction) ** 2,5)

    print(f"Iteration {iteration}")
    print(f"- weight: {weight}")
    print(f"- Prediction: {prediction}; Error: {error};")
    print(f"- up_prediction: {up_prediction}; up_error: {up_error};")
    print(f"- down_prediction: {down_prediction}; down_error: {down_error};")
    print()

    do_iteration = (error != 0.00000)
    iteration += 1

    if(down_error < up_error):
        weight = weight - step_amount

    if(down_error > up_error):
        weight = weight + step_amount
```
