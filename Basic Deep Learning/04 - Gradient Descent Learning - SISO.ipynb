```python
MAX_ITERATION = 100
MINUTES_RUNNING = 20
ACTUAL_CALORIES_BURNED = 180
INITIAL_WEIGHT = 7

SLOW_LEARNING_RATE = 0.001
OSCILLATING_LEARNING_RATE = 0.1

learning_rate = SLOW_LEARNING_RATE


def neural_network(input, weight):
    predicted_value = input * weight
    return predicted_value


input = MINUTES_RUNNING
expected_value = ACTUAL_CALORIES_BURNED
weight = INITIAL_WEIGHT
iteration = 1

while True:
    prediction = neural_network(input, weight)
    error = (prediction - ACTUAL_CALORIES_BURNED) ** 2
    error_rounded = round(error, 5)
    error_derivative = 2 * input * (prediction - expected_value)
    weight_adjustment = learning_rate * error_derivative
    weight_new = weight - weight_adjustment

    print(f"=================== Iteration: {iteration}")
    print(f"input             : {input}")
    print(f"expected_value    : {expected_value}")
    print()
    print(f"weight            : {weight}")
    print(f"prediction        : {prediction}")
    print()
    print(f"error             : {error}")
    print(f"error (round)     : {error_rounded} {"     ===> Yeayyyy the error is zero !!!!" if error_rounded == 0 else ""}")
    print(f"error_derivative  : {error_derivative}")
    print()
    print(f"learning_rate     : {learning_rate}")
    print(f"weight_adjustment : {weight_adjustment}")    
    print(f"weight_new        : {weight_new}")
    print()

    iteration += 1
    weight = weight_new
    if(error_rounded == 0 or iteration > MAX_ITERATION):
        break
```
