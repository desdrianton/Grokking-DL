```python
input = 0.2
expected_value = 8
weight = 10
learning_rate = 18

iteration = 1
has_error = True


def dot_product(input, weight):
    predicted_value = input * weight
    return predicted_value


def print_info():
    print(f"=================== Iteration: {iteration}")
    print("SETUP")
    print(f"- expected_value    : {expected_value}")
    print(f"- learning_rate     : {learning_rate}")
    print(f"- input             : {input}")
    print()
    print("PREDICTING")
    print(f"- weight            : {weight}")
    print(f"- nn_prediction     : {nn_prediction}")
    print()
    print("CALCULATING ERROR")
    print(f"- error_raw         : {error_raw}")
    print(f"- error             : {error}")
    print()
    print("ADJUSTING WEIGHT / LEARNING")
    print(f"- error_derivative  : {error_derivative}")
    print(f"- weight_adjustment : {weight_adjustment}")    
    print(f"- weight_updated    : {weight_updated}")
    print()


while has_error:
    # Predicting
    nn_prediction = dot_product(input, weight)

    # Calculating error
    error_raw = nn_prediction - expected_value
    error = error_raw ** 2

    # Adjusting weight / learning
    error_derivative = 2 * input * error_raw
    weight_adjustment = learning_rate * error_derivative
    weight_updated = weight - weight_adjustment

    # Still need learning ?
    has_error = (round(error, 5) != 0)
  

    print_info()
    iteration += 1
    weight = weight_updated
    
    # Circuit breaker
    if(iteration > 100000):
        print()
        print("!!!!! Reaching circuit breaker point !!!!!")
        break
```
