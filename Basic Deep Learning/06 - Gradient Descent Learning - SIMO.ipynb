```python
input = 0.2
expected_values = [8, 46, 0.1]
weights = [10, 20, 3]

learning_rate = 24
iteration = 1
has_error = True


def dot_product(input, weights):
    predicted_values = []

    for weight in weights:
        predicted_values.append(input * weight)
    
    return predicted_values


def print_info():
    print(f"=================== Iteration: {iteration}")
    print("SETUP")
    print(f"- expected_value     : {expected_values}")
    print(f"- learning_rate      : {learning_rate}")
    print(f"- input              : {input}")
    print()
    print("PREDICTING")
    print(f"- weight             : {weights}")
    print(f"- nn_prediction      : {nn_predictions}")
    print()
    print("CALCULATING ERROR")
    print(f"- error_raws         : {error_raws}")
    print(f"- errors             : {errors}")
    print(f"- total_error        : {total_error}")
    print()
    print("ADJUSTING WEIGHT / LEARNING")
    print(f"- error derivatives  : {errors_derivative}")
    print(f"- weight_adjustments : {weights_adjustment}")    
    print(f"- weight_news        : {weights_updated}")
    print()


while has_error:
    # Predicting
    nn_predictions = dot_product(input, weights)

    # Calculating error
    error_raws = []
    for nn_prediction, expected_value in zip(nn_predictions, expected_values):
        error_raws.append(nn_prediction - expected_value)
    
    errors = []
    for error_raw in error_raws:
        errors.append(error_raw ** 2)
    

    # Adjusting weight / learning
    errors_derivative = []
    for error_raw in error_raws:
        errors_derivative.append(2 * input * error_raw)
    
    weights_adjustment = []
    for error_derivative in errors_derivative:
        weights_adjustment.append(learning_rate * error_derivative)

    weights_updated = []
    for weight, weight_adjustment in zip(weights, weights_adjustment):
        weights_updated.append(weight - weight_adjustment)
    

    # Still need learning ?
    total_error = 0
    for error in errors:
        total_error += error
    total_error = round(total_error, 5) 
    has_error = (total_error != 0)
  

    print_info()
    iteration += 1
    weights = weights_updated
    
    # Circuit breaker
    if(iteration > 100000):
        print()
        print("!!!!! Reaching circuit breaker point !!!!!")
        break
```
