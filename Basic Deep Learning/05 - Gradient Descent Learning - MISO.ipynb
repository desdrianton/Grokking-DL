```python
inputs = [0.2, 4, 0.1]
expected_value = 8
weights = [0, 0, 0]
learning_rate = 0.01
iteration = 1
has_error = True


def dot_product(inputs, weights):
    assert(len(inputs) == len(weights))

    predicted_value = 0
    for input, weight in zip(inputs, weights):
        predicted_value += input * weight

    return predicted_value


def print_info():
    print(f"=================== Iteration: {iteration}")
    print("SETUP")
    print(f"- expected_value     : {expected_value}")
    print(f"- learning_rate      : {learning_rate}")
    print(f"- inputs             : {inputs}")
    print()
    print("PREDICTING")
    print(f"- weights            : {weights}")
    print(f"- nn_prediction      : {nn_prediction}")
    print()
    print("CALCULATING ERROR")
    print(f"- error_raw          : {error_raw}")
    print(f"- error              : {error}")
    print()
    print("ADJUSTING WEIGHT / LEARNING")
    print(f"- errors_derivative  : {errors_derivative}")
    print(f"- weights_adjustment : {weights_adjustment}")    
    print(f"- weights_new        : {weights_updated}")
    print()


while has_error:
    # Predicting
    nn_prediction = dot_product(inputs, weights)

    # Calculating error
    error_raw = nn_prediction - expected_value
    error = error_raw ** 2
    
    # Adjusting weight / learning
    assert(len(inputs) == len(weights))

    errors_derivative = []
    weights_adjustment = []
    weights_updated = []
    for i in range(len(inputs)):
        error_derivative = 2 * inputs[i] * error_raw
        errors_derivative.append(error_derivative)

        weight_adjustment = learning_rate * error_derivative
        weights_adjustment.append(weight_adjustment)

        weight_updated = weights[i] - weight_adjustment
        weights_updated.append(weight_updated)

    # Still need learning ?
    has_error = (round(error, 5) != 0)


    print_info()
    iteration += 1
    weights = weights_updated

    # Circuit breaker
    if(iteration > 100000):
        print()
        print("!!!!! Reaching circuit breaker point !!!!!")
        break
```
