```python
inputs = [0.2, 4, 0.1]
expected_value = 8
weights = [0, 0, 0]
learning_rate = 0.01
iteration = 1
has_error = True


def dot_product(inputs, weights):
    assert(len(inputs) == len(weights))

    predicted_value = 0
    for input, weight in zip(inputs, weights):
        predicted_value += input * weight

    return predicted_value


def print_info():
    print(f"=================== Iteration: {iteration}")
    print(f"input               : {inputs}")
    print(f"expected_value      : {expected_value}")
    print()
    print(f"weights             : {weights}")
    print(f"nn_prediction       : {nn_prediction}")
    print()
    print(f"error_mse           : {error_mse}")
    print(f"error_mse_derivative: {error_mse_derivative}")
    print()
    print(f"learning_rate       : {learning_rate}")
    print(f"weights_new         : {weights_new}")
    print()


while has_error:
    # Predicting
    nn_prediction = dot_product(inputs, weights)

    # Calculating error
    error_raw = nn_prediction - expected_value
    error_mse = error_raw ** 2
    
    # Adjusting weight
    assert(len(inputs) == len(weights))

    weights_new = []
    for i in range(len(inputs)):
        error_mse_derivative = 2 * inputs[i] * error_raw
        weight_adjustment = learning_rate * error_mse_derivative
        weight_new = weights[i] - weight_adjustment

        weights_new.append(weight_new)

    # Still need learning ?
    has_error = (round(error_mse, 5) != 0)


    print_info()
    iteration += 1
    weights = weights_new

    # Circuit breaker
    if(iteration > 100000):
        print()
        print("!!!!! Reaching circuit breaker point !!!!!")
        break
```
