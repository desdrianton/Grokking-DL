```python
MAX_ITERATION = 100
MINUTES_RUNNING = 20
ACTUAL_CALORIES_BURNED = 180
INITIAL_WEIGHT = 7

SLOW_LEARNING_RATE = 0.1            # 21 iterations
FASTEST_LEARNING_RATE = 2           # Only 2 iterations
OSCILLATING_LEARNING_RATE = 3       # > 100 iterations

learning_rate = SLOW_LEARNING_RATE


def neural_network(input, weight):
    predicted_value = input * weight
    return predicted_value


input = MINUTES_RUNNING
expected_value = ACTUAL_CALORIES_BURNED
weight = INITIAL_WEIGHT
iteration = 1

while True:
    prediction = neural_network(input, weight)
    error = (prediction - ACTUAL_CALORIES_BURNED) ** 2

    weight_upward = weight + learning_rate
    prediction_upward = neural_network(input, weight_upward)
    error_upward = (prediction_upward - ACTUAL_CALORIES_BURNED) ** 2

    weight_downwards = weight - learning_rate
    prediction_downwards = neural_network(input, weight_downwards)
    error_downwards = (prediction_downwards - ACTUAL_CALORIES_BURNED) ** 2

    print(f"======================== Iteration: {iteration}")
    print(f"weight                 : {weight}")
    print(f"prediction             : {prediction}")
    print(f"error                  : {error}")
    print(f"error (round)          : {round(error, 5)} {"     ===> Yeayyyy the error is zero !!!!" if round(error, 5) == 0 else ""}")
    print()
    print(f"weight_upward          : {weight_upward}")
    print(f"prediction_upward      : {prediction_upward}")
    print(f"error_upward           : {error_upward}")
    print(f"error_upward (round)   : {round(error_upward, 5)}")
    print()
    print(f"weight_downwards       : {weight_downwards}")
    print(f"prediction_downwards   : {prediction_downwards}")
    print(f"error_downwards        : {error_downwards}")
    print(f"error_downwards (round): {round(error_downwards, 5)}")
    print()


    iteration += 1
    if(round(error, 5) == 0 or iteration > MAX_ITERATION):
        break

    if(error_upward < error_downwards):
        weight = weight_upward
        print(f"new weight using weight_upward: {weight}\n")
    else:
        weight = weight_downwards
        print(f"new weight using weight_downward: {weight}\n")
```
