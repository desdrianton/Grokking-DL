```python
import numpy as np
```

```python
# 1. Setup
input = np.array([[0.2]])
initial_weight = np.array([[10]])
expected_value = np.array([[8]])
alpha = 18 # Learning rate
need_learning = True
weight = initial_weight
iteration = 1

while need_learning:
    # 2. NN Prediction
    predicted_value = input.dot(weight)

    # 3. Calculate Error
    delta = predicted_value - expected_value
    error = delta ** 2

    # 4. Weight Adjustment
    error_derivative = 2 * input * delta
    weight_adjustment = alpha * error_derivative
    new_weight = weight - weight_adjustment

    # 5. Still need learning?
    need_learning = (round(error[0][0], 5) != 0)

    # Print information
    print(f"==================== Iteration {iteration}")
    print("Setup")
    print(f"- Input             : {input}")
    print(f"- expected_value    : {expected_value}")
    print(f"- alph              : {alpha}")
    print("NN Prediction")
    print(f"- weight            : {weight}")
    print(f"- predicted_value   : {predicted_value}")
    print("Calculate Error")
    print(f"- delta             : {delta}")
    print(f"- error             : {error}")
    print("Weight Adjustment")
    print(f"- error_derivative  : {error_derivative}")
    print(f"- weight_adjustment : {weight_adjustment}")
    print(f"- new_weight        : {new_weight}")
    print("Need Learning")
    print(f"- need_learning     : {need_learning}")
    print()

    # Update weight for next iteration
    weight = new_weight

    # Circuit breaker (max 1000 iteration)
    iteration = iteration + 1
    if(iteration > 1000):
        break
```
